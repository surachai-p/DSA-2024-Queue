### ให้เพิ่มเมธอด `clear()` ในคลาส Queue เพื่อล้างข้อมูลทั้งหมดในคิว

````python
class Queue:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None
    
    def front(self):
        if not self.is_empty():
            return self.items[0]
        return None
    
    def size(self):
        return len(self.items)
    
    def display(self):
        print(f"Queue: {self.items}")

    def clear(self):
        self.items = []  # ล้างข้อมูลทั้งหมดในคิว

# ทดสอบการใช้งานเมธอด clear()
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.display()  # ควรได้ [1, 2, 3]

q.clear()
q.display()  # ควรได้ [] (คิวว่าง)
````

### ให้เพิ่มเมธอด `peek_last()` ในคลาส Queue เพื่อดูข้อมูลตัวสุดท้ายในคิวโดยไม่นำออก

````python
class Queue:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None
    
    def front(self):
        if not self.is_empty():
            return self.items[0]
        return None
    
    def size(self):
        return len(self.items)
    
    def display(self):
        print(f"Queue: {self.items}")

    def clear(self):
        self.items = []  # ล้างข้อมูลทั้งหมดในคิว

    def peek_last(self):
        if not self.is_empty():
            return self.items[-1]
        return None

# ทดสอบการใช้งานเมธอด clear() และ peek_last()
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.display()  # ควรได้ [1, 2, 3]

print(q.peek_last())  # ควรได้ 3

q.clear()
q.display()  # ควรได้ [] (คิวว่าง)
````

### ตัวอย่างการประยุกต์ใช้คิวในการจำลองการรอคิวร้านอาหาร จะเกิดความผิดพลาดในการทำงานกรณีใด จงอธิบาย และจะแก้ไข Code อย่างไรเพื่อแก้ไขปัญหาดังกล่าว (สร้างไฟล์ใหม่ ตั้งขื่อ UpdateQueueRestaurant.py)

1. ตรวจสอบความถูกต้องของข้อมูล มีการเพิ่มฟังก์ชันสำหรับจัดการกับลูกค้าที่ยกเลิกการรอคิว
2. หากมีการเพิ่มหรือลบลูกค้าในขณะที่กำลังดำเนินการเสิร์ฟ อาจทำให้ลำดับคิวผิดเพี้ยนได้

````python
class Queue:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None
    
    def front(self):
        if not self.is_empty():
            return self.items[0]
        return None
    
    def size(self):
        return len(self.items)
    
    def display(self):
        print(f"Queue: {self.items}")

class Restaurant:
    def __init__(self):
        self.waiting_queue = Queue()
        
    def add_customer(self, name):
        if name in self.waiting_queue.items:
            print(f"ลูกค้า {name} อยู่ในคิวแล้ว")
        else:
            self.waiting_queue.enqueue(name)
            print(f"ลูกค้า {name} เข้าคิวที่ {self.waiting_queue.size()}")
        
    def serve_customer(self):
        if not self.waiting_queue.is_empty():
            customer = self.waiting_queue.dequeue()
            print(f"กำลังเสิร์ฟลูกค้า: {customer}")
        else:
            print("ไม่มีลูกค้าในคิว")
            
    # ฟังก์ชันเพิ่มสำหรับยกเลิกลูกค้า
    def cancel_order(self, name):
        if name in self.waiting_queue.items:
            self.waiting_queue.items.remove(name)
            print(f"ลูกค้า {name} ยกเลิกการรอคิว")
        else:
            print(f"ไม่พบลูกค้า {name} ในคิว")
            
    def show_queue(self):
        print(f"คิวปัจจุบัน: {self.waiting_queue.items}")
        print(f"จำนวนลูกค้าในคิว: {self.waiting_queue.size()}")

# สร้างร้านอาหาร
restaurant = Restaurant()

# เพิ่มลูกค้าเข้าคิว
restaurant.add_customer("อานนท์")
restaurant.add_customer("บุญมี")
restaurant.add_customer("จินดา")

# แสดงสถานะคิว
restaurant.show_queue()

# เสิร์ฟลูกค้า 2 คน
restaurant.serve_customer()
restaurant.serve_customer()

# แสดงสถานะคิวอีกครั้ง
restaurant.show_queue()
````

4. ประยุกต์ใช้คิวในการจำลองระบบคิวธนาคาร โดยมีฟังก์ชันการทำงานดังนี้:
   - เพิ่มลูกค้าเข้าคิว พร้อมประเภทธุรกรรม (ฝาก/ถอน/ชำระบิล)
   - เรียกลูกค้าเข้ารับบริการ
   - แสดงจำนวนคิวที่รอ
   - แสดงประมาณการเวลารอ (สมมติว่าธุรกรรมใช้เวลาเฉลี่ย 5 นาที/คน)

````python
class Bank:
    def __init__(self):
        self.waiting_queue = []  # เปลี่ยนเป็น list เพื่อเก็บ tuple (ชื่อ, ประเภทธุรกรรม)
        self.average_transaction_time = 5  # นาที/คน

    def add_customer(self, name, transaction_type):
        self.waiting_queue.append((name, transaction_type))
        print(f"ลูกค้า {name} เข้าคิวที่ {len(self.waiting_queue)} ด้วยธุรกรรม {transaction_type}")

    def serve_customer(self):
        if self.waiting_queue:
            customer, transaction_type = self.waiting_queue.pop(0)
            print(f"กำลังเสิร์ฟลูกค้า: {customer} (ธุรกรรม: {transaction_type})")
        else:
            print("ไม่มีลูกค้าในคิว")

    def show_queue(self):
        print("คิวปัจจุบัน:")
        for i, (name, transaction_type) in enumerate(self.waiting_queue, 1):
            print(f"{i}. {name} ({transaction_type})")
        print(f"จำนวนลูกค้าในคิว: {len(self.waiting_queue)}")

    def estimated_waiting_time(self):
        if self.waiting_queue:
            num_customers = len(self.waiting_queue)
            estimated_time = num_customers * self.average_transaction_time
            print(f"ประมาณการเวลาที่ต้องรอ: {estimated_time} นาที")
        else:
            print("ไม่มีลูกค้าในคิว")

# สร้างธนาคาร
bank = Bank()

# เพิ่มลูกค้าเข้าคิว
bank.add_customer("อานนท์", "ฝาก")
bank.add_customer("บุญมี", "ถอน")
bank.add_customer("จินดา", "ชำระบิล")

# แสดงสถานะคิว
bank.show_queue()

# ประมาณการเวลาที่ต้องรอ
bank.estimated_waiting_time()

# เสิร์ฟลูกค้า 2 คน
bank.serve_customer()
bank.serve_customer()

# แสดงสถานะคิวอีกครั้ง
bank.show_queue()
````

5. จงเขียนโปรแกรมจำลองระบบจัดการคิวร้านตัดผม โดยมีความสามารถดังนี้:
   - เพิ่มลูกค้าเข้าคิว พร้อมระบุบริการที่ต้องการ (ตัดผม, สระ, ย้อมสี)
   - แสดงรายชื่อลูกค้าที่รออยู่ในคิว
   - เรียกลูกค้าคนถัดไป
   - แสดงเวลารอโดยประมาณ (ตัดผม 30 นาที, สระ 20 นาที, ย้อมสี 60 นาที)

````python
class Customer:
    def __init__(self, name, service):
        self.name = name
        self.service = service
        self.estimated_time = self.get_estimated_time()

    def get_estimated_time(self):
        if self.service == "ตัดผม":
            return 30
        elif self.service == "สระ":
            return 20
        elif self.service == "ย้อมสี":
            return 60
        else:
            return 0

class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, customer):
        self.items.append(customer)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None

    def display(self):
        if not self.is_empty():
            for customer in self.items:
                print(f"ชื่อ: {customer.name}, บริการ: {customer.service}, เวลาประมาณ: {customer.estimated_time} นาที")
        else:
            print("คิวว่าง")

# สร้างคิว
queue = Queue()

# เพิ่มลูกค้าเข้าคิว
queue.enqueue(Customer("สมชาย", "ตัดผม"))
queue.enqueue(Customer("นางสาวบุญมี", "สระ"))
queue.enqueue(Customer("นายสิทธา", "ย้อมสี"))

# แสดงคิว
queue.display()

# เรียกลูกค้าคนถัดไป
customer = queue.dequeue()
if customer:
    print(f"เรียกลูกค้า: {customer.name}")
else:
    print("ไม่มีลูกค้าในคิว")
````

1. Queue ต่างจาก Stack อย่างไร?
Stack: โครงสร้างข้อมูลแบบ LIFO (Last In, First Out) หมายถึง ข้อมูลที่เพิ่มเข้ามาล่าสุดจะถูกนำออกก่อน เช่น การซ้อนจาน
Queue: โครงสร้างข้อมูลแบบ FIFO (First In, First Out) หมายถึง ข้อมูลที่เข้ามาก่อนจะถูกนำออกก่อน เช่น การต่อแถวซื้อของ

2. เพราะเหตุใดการ dequeue จึงใช้ pop(0) แทนที่จะเป็น pop()?
pop(0): ลบข้อมูลตัวแรกในลำดับ (index 0) ซึ่งเหมาะสำหรับการทำ Queue เพราะต้องการลบข้อมูลที่เข้ามาก่อนสุด
pop(): ลบข้อมูลตัวสุดท้ายในลำดับ (index -1) ซึ่งเหมาะสำหรับการทำ Stack
การ dequeue ใช้ pop(0) เพื่อให้สอดคล้องกับลักษณะของ Queue ที่ทำงานแบบ FIFO

3. ยกตัวอย่างการประยุกต์ใช้ Queue ในชีวิตประจำวันมา 3 ตัวอย่าง
